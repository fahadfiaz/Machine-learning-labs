### Importing libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from string import ascii_lowercase
from numpy.linalg import inv
from math import sqrt
%matplotlib inline
```

### loading Data

```python
#reading data files
store_df=pd.read_csv("store.csv")
train_df=pd.read_csv("train.csv")
```

### marging 2 dataframes which will help us later

```python
#Merging both the Dataframes into one based on the "Store" ID
df=store_df.merge(train_df,on=["Store"],how="inner")
```

### sorting data in ascending order  based on date

```python
df2 = df.sort_values('Date',ascending=True)
```

### aranging data such that I have store on x-asix and days on y-axis

```python
c=1
dict1 = {}
for i in range(1,1116):
    dict1[c] = df2[df2["Store"]==i]["Sales"].values
    c=c+1
    
df_3 = pd.DataFrame.from_dict(dict1,orient='index')
df_4 = df_3.T
df_4.index = df_4.index + 1
df_5 = df_4.T
df_5
```

```python
df_5.shape
```

## B,C .)Test train split as described in exercise, size of sets

```python
X_train = df_5.iloc[:1000,:900]
Y_train = df_5.iloc[:1000,900:]

X_test = df_5.iloc[1000:,:900]
Y_test = df_5.iloc[1000:,900:]
```

### Fillling nan values with 0

```python
X_train.fillna(0,inplace=True)
Y_train.fillna(0,inplace=True)
X_test.fillna(0,inplace=True)
Y_test.fillna(0,inplace=True)
```

## D.) learning beta_hat on X_train using Y_train

```python
X_transpose = X_train.T
A = np.dot(X_transpose , X_train)

# beta_All = []
beta_All = np.zeros(shape=(42,900))
for i in range(42):
    b = np.dot(X_transpose , Y_train.iloc[:,i])
    inverse_A = inv(A)
    beta_hat = np.dot(inverse_A,b)
    beta_All[i] = beta_hat
```

```python
beta_All.shape
```

### Doing prediction using X_train and beta_all

```python
predictions = np.dot(X_train,beta_All.T)
```

```python
predictions.shape
```

### Saving errors on X_train

```python
erros = np.zeros(shape=(42,2))
```

```python
for i in range(42):
    numerator = ((Y_train.iloc[:,i] - predictions[:,i])**2).sum()
    rmse = sqrt((numerator)/predictions.shape[0])
    erros[i][0] = rmse
    
    numerator_mae = (np.absolute(Y_train.iloc[:,i] - predictions[:,i])).sum()
    mae = numerator_mae/predictions.shape[0]
    erros[i][1] = mae
error_df = pd.DataFrame({'RMSE': erros[:, 0], 'MAE': erros[:, 1]})
error_df.index = error_df.index + 1
```

```python
error_df ## error of next 42 days
```

```python
print(f'RMSE of all next 42 days: {error_df["RMSE"].sum()} on X_train')
print(f'MAE of all next 42 days: {error_df["MAE"].sum()} on X_train')
```

### Doing prediction on X_test using beta_all learned previously. 

```python
predictions_Test = np.dot(X_test,beta_All.T)
```

### Saving errors on X_test

```python
error_test_set= np.zeros(shape=(42,2))
```

```python
for i in range(42):
    numerator = ((Y_test.iloc[:,i] - predictions_Test[:,i])**2).sum()
    rmse = sqrt((numerator)/predictions_Test.shape[0])
    error_test_set[i][0] = rmse
    
    numerator_mae = (np.absolute(Y_test.iloc[:,i] - predictions_Test[:,i])).sum()
    mae = numerator_mae/predictions_Test.shape[0]
    error_test_set[i][1] = mae
error_test_df = pd.DataFrame({'RMSE': error_test_set[:, 0], 'MAE': error_test_set[:, 1]})
error_test_df.index = error_test_df.index+1
```

```python
error_test_df ## error of next 42 days using X_Test
```

```python
print(f'RMSE of all next 42 days: {error_test_df["RMSE"].sum()} on X_test')
print(f'MAE of all next 42 days: {error_test_df["MAE"].sum()} on X_test')
```

### sorting dataframe by row

```python
df_c = df2[["Store","Sales","Date"]].sort_values(by=['Store','Date'],ascending=True)
```

### add new coloumn in dataframe which is day name in date coloumn

```python
df_c['Day'] = pd.DatetimeIndex(df_c['Date']).day_name()
df_c.head(n=3)
```

### groupby first on day, then on store to get sale of each store for each day in week

```python
sales_gc = df_c.groupby(['Day','Store']).mean()
```

```python
# resetting index to convert multi index to single index
rsd = sales_gc.reset_index()  
```

```python
rsd
```

### For each store get its store id and day name. Then use  same store_id and day_name to get sales of that day from rsd dataframe (made in previous step) and save these sales in dictionary with id same as storeId. Each item of dictionary is store id and mean of all sales recorded for that day of week in the past

```python
c=1
dict1 = {}
for i in range(1,1116):    
    sds = df_c[df_c["Store"]==i]
#     datess = df_c[df_c["Store"]==i]["Day"]
    datess = df_c[(df_c["Store"]==i) & (df_c["Day"] != "Sunday")]["Day"]
    
    mean_of_sales = []
    for k in range(datess.shape[0]):
        day_cur = datess.iloc[k]
        sale_on_that_day = rsd[(rsd["Store"] == i) & (rsd["Day"] ==day_cur)]
        mean_of_sales.append(sale_on_that_day.values[0][-1])
    
    
    dict1[c] = mean_of_sales
    c= c+1
    print("Done Store: ",i)
```

### converting dictionary to dataframe

```python
df_fin = pd.DataFrame.from_dict(dict1,orient='index')
df_4_fin = df_fin.T
df_4_fin.index = df_4_fin.index + 1
df_5_fin = df_4_fin.T
df_5_fin
```

### H.)


<p>Linear regression is not a good choice for this task because this task is classicial example for Auto-regression. In this case, prediction is time dependent. To predict next day sales, model should take into account previous predictions. In auto regression case, model Y is based on previous values of Y. Our example is similar to this where we are predicting sales of next 42 days. Our model should take into considerations the previous predicted sales. </p>
    
<p>In case of linear regression, the prediction is independent of time and it does not take into account previous predictions.In this case, model Y is based on distribution of X</p>

<p>so for this task linear regresssion is a bad choice</p>
